問題一

- 讀入 excel 試算表：https://storage.googleapis.com/py_ml_datasets/Energy%20Indicators.xls 並儲存成 DataFrame，命名為 energy

   - 擷取這四個變數並重新命名：['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']
   - 將 Energy Supply 單位改為 gigajoules(1 petajoule = 1,000,000 gigajoules)
   - 將 ... 的資料記錄為遺漏值 np.NaN重新命名國家名稱：
      - "Republic of Korea": "South Korea"
      - "United States of America": "United States"  
      - "United Kingdom of Great Britain and Northern Ireland": "United Kingdom"
      - "China, Hong Kong Special Administrative Region": "Hong Kong"
      - 移除小括號或數字，像是 'Bolivia (Plurinational State of)': 'Bolivia'、'Switzerland17': 'Switzerland'

```python
import pandas as pd
xls_url = "https://storage.googleapis.com/py_ml_datasets/Energy%20Indicators.xls"
energy = pd.read_excel(xls_url)
energy = energy.iloc[17:244,2:6]
energy.columns = ['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']  
energy = energy.reset_index()
energy["Energy Supply"] = energy["Energy Supply"][:]*1000000
```

```python
country_list = list(energy["Country"])
for i in country_list:
    if "Republic of Korea" in i:
        energy["Country"] = energy["Country"].replace(i,"South Korea")
    if "United States of America" in i:
        energy["Country"] = energy["Country"].replace(i,"United States")
    if "United Kingdom of Great Britain and Northern Ireland" in i:
        energy["Country"] = energy["Country"].replace(i,"United Kingdom")
    if "China, Hong Kong Special Administrative Region" in i:
        energy["Country"] = energy["Country"].replace(i,"Hong Kong")
        
energy['Country'] = energy['Country'].str.replace("[0-9]+$", "") # 正則表達式: $表示字符串以此結尾。
energy['Country'] = energy['Country'].str.replace("\s\(.+\)$", "") # \s 空白字元，\S非空白字元
energy.head()
```

```python
energy[energy['Country'] == 'South Korea'] # South Korea有重複的兩筆資料 
```

- 讀入 csv 檔：https://storage.googleapis.com/py_ml_datasets/world_bank.csv 並儲存成 DataFrame，命名為 GDP
- 重新命名國家名稱：
   - "Korea, Rep.": "South Korea" 
   - "Iran, Islamic Rep.": "Iran"
   - "Hong Kong SAR, China": "Hong Kong"

```python
csv_url = "https://storage.googleapis.com/py_ml_datasets/world_bank.csv"
GDP = pd.read_csv(csv_url,skiprows = 4)
gdp_country_list = list(GDP["Country Name"])
for i in gdp_country_list:
    if "Korea, Rep." in i:
        GDP["Country Name"] = GDP["Country Name"].replace(i,"South Korea")
    if "Iran, Islamic Rep" in i:
        GDP["Country Name"] = GDP["Country Name"].replace(i,"Iran")
    if "Hong Kong SAR, China" in i:
        GDP["Country Name"] = GDP["Country Name"].replace(i,"Hong Kong")
GDP.head()
```

- 讀入 excel 試算表：https://storage.googleapis.com/py_ml_datasets/scimagojr-3.xlsx 並儲存成 DataFrame，命名為 ScimEn

```python
xlsx_url = "https://storage.googleapis.com/py_ml_datasets/scimagojr-3.xlsx"
ScimEn = pd.read_excel(xlsx_url)
ScimEn.head(16)
```

- 利用國家名稱將上述三個 DataFrame （GDP、Energy、ScimEn）聯結
- 只需要 2006-2015 年的 GDP 資料以及 ScimEn 中排名前 15 的國家
- 資料框的索引值為國家名稱
- 資料框的變數名稱為：['Rank', 'Documents', 'Citable documents', 'Citations', 'Self-citations', 'Citations per document', 'H index', 'Energy Supply', 'Energy Supply per Capita', '% Renewable', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015']

```python
energy = energy.set_index("Country")
GDP = GDP.set_index("Country Name")
ScimEn = ScimEn.set_index("Country")
```

```python
outer_joined_df = energy.join(ScimEn, how = 'outer') 
outer_joined_df = outer_joined_df.join(GDP.loc[:, '2006':'2015'], how = 'outer')
```

```python
outer_joined_df_top15 = outer_joined_df[outer_joined_df['Rank']<=15]
outer_joined_df_top15 = outer_joined_df_top15[['Rank', 'Documents', 'Citable documents', 'Citations',
                                     'Self-citations', 'Citations per document', 'H index',
                                     'Energy Supply', 'Energy Supply per Capita', '% Renewable',
                                     '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015']]
outer_joined_df_top15
```
