# PyPlot

- `plt.scatter()`: 散佈圖（numeric ~ numeric）

```python
import matplotlib.pyplot as plt
import pandas as pd

gapminder = pd.read_csv("https://storage.googleapis.com/learn_pd_like_tidyverse/gapminder.csv")
gapminder.shape
```

```python
gapminder.head()
```

```python
# scatter plot
x_arr = gapminder["gdpPercap"].values
y_arr = gapminder["lifeExp"].values
plt.scatter(x_arr, y_arr)
plt.show()
```

```python
# scatter plot
x_arr = gapminder["gdpPercap"].values
y_arr = gapminder["lifeExp"].values
plt.scatter(x_arr, y_arr, c = "red") # 把顏色變成紅色
plt.title("Gapminder Viz") # 增加標題
plt.xlabel("GDP Per Capita") # 增加 x 軸名稱
plt.ylabel("Life Expectancy") # 增加 y 軸名稱
plt.show()
```

```python
'''
要訂顏色可以使用字串直接指定，也可以用color hex，詳細對應請至下方網站搜尋
https://www.color-hex.com/
'''  
# scatter plot
colors = ["red", "green", "blue", "#81811D", "#1D6D81"]
for cont, color in zip(gapminder["continent"].unique(), colors):
    cont_df = gapminder[gapminder["continent"] == cont]
    x_arr = cont_df["gdpPercap"].values
    y_arr = cont_df["lifeExp"].values
    plt.scatter(x_arr, y_arr, c=color, label=cont)
plt.title("Gapminder Viz")
plt.xlabel("GDP Per Capita")
plt.ylabel("Life Expectancy")
plt.legend()
plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(10, 6))

# scatter plot
colors = ["red", "green", "blue", "#81811D", "#1D6D81"]
n_rows = [0, 0, 0, 1, 1]
n_cols = [0, 1, 2, 0, 1]
for cont, color, n_row, n_col in zip(gapminder["continent"].unique(), colors, n_rows, n_cols):
    cont_df = gapminder[gapminder["continent"] == cont]
    x_arr = cont_df["gdpPercap"].values
    y_arr = cont_df["lifeExp"].values
    axes[n_row, n_col].scatter(x_arr, y_arr, c=color, label=cont)
    axes[n_row, n_col].set_xticks([])
    axes[n_row, n_col].set_title(cont)
axes[1, 2].set_visible(False)
plt.tight_layout()
plt.show()
```
